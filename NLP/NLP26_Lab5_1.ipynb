The following function *tokenbuilder* tokenizes text.
1. Study the function and use it with and without the stemmer.
2. What is the role of "yield"?

```python

import string
from nltk.stem.porter import *
from nltk.stem.snowball import SnowballStemmer
def tokenbuilder(txt):
    """
      Take text, lowercase it, tokenize it, drop punctuation/noisy symbols,
      mark periods as sentence ends, and emit the cleaned tokens one by one.

      Args:
        txt(string): text to be preprocessed and tokenized text.

      Return:
        tokenized text.
    """
    #stem = SnowballStemmer("english", ignore_stopwords=True)
    stem = PorterStemmer( mode = 'NLTK_EXTENSIONS')
    txt = txt.lower()
    #pat = r"\W"
    #txt_tokens = re.split(pat, txt)
    rmv = ["''", "'", "'s", "``", "`", "--", "__"]
    for token in nltk.word_tokenize(txt):
        if token in string.punctuation or token in rmv:
            if (token == r'.'):
                token = r'</s>'
            else:
                continue
        #yield stem.stem(token)
        yield token
```

```python
from collections import defaultdict
def vectorize(txt):
    """
      Method does a count of the occurence of each word in the text.

      Args:
        txt(string): text to be preprocessed and tokenized text.

      Return:
        dictionary of tokens and their counts.
    """
    features = defaultdict(int)
    for token in tokenbuilder(txt):
        features[token] += 1
    return features
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stopWords = set(stopwords.words('english'))
#stopWords
```

```python
#the following function is completed, it creates list of tokens
def tokenList(txt, stp = 0):
    """
      Builds a token list (starting with <b>), optionally filtering out stopwords
      based on a flag.

      Args:
        txt(str): text to be preprocessed and converted into tokens.
        stp(int): flag for filtering out stopwords.

      Return:
        list of tokens.
    """
    toke_list = [r'<b>']
    if (stp == 0):
        for token in tokenbuilder(txt):
            toke_list.append(token)
    else:
         for token in tokenbuilder(txt):
                if not token in stopWords:
                    toke_list.append(token)
    return toke_list
```

#### Unigrams    
In the function below you will build a unigram as a dictionary.
1. The input is a text string or file.
2. The function uses *tokenbuilder*() to update the dictionary. The keys are tokens and the values the counts of the tokens.


```python
#build the unigram as a dictionary
#The keyword argument is set to true for a file. Otherwise it is a string.
def unigram(txt, fl=True):
    """
      Builds a word-frequency (unigram) dictionary from a text or a file
      using cleaned tokens.

      Args:
        txt(str): input text or file path.
        f1(bool): if True, treat txt as a file path.

      Return:
        dict: dictionary maping of tokens to their counts.
    """
    unigm = defaultdict(int)
    if fl:
        with open(txt, encoding='utf-8') as f:
            txt = f.read()
# Use the tokenbuilder function to create the dictionary.
    for tok in tokenbuilder(txt):
        unigm[tok] +=1
    pass
    return unigm
```

### Bigrams

```python
#Exercise
#build the bigram as a dictionary
def bigram(txt, fl=True):
    """
      Builds a dictionary counting how often each pair of consecutive tokens
      appears in the text.

      Args:
        txt(str): input text or file path.
        f1(bool): if True, treat txt as a file path.

      Returns:
        dict: dictionary mapping of bigrams to their counts.
    """
    bigm = defaultdict(int)
    if fl:
         with open(txt, encoding='utf-8') as f:
            txt = f.read()
    tl = list(tokenbuilder(txt))
    for i in range(len(tl)-1):
        bigm[(tl[i], tl[i+1])] +=1
    return bigm
```
